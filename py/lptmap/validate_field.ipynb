```python
import numpy as np 
import healpy as hp
import matplotlib.pyplot as plt
import skytools as st
```

```python
hpx_datapath = '/Users/shamik/opt/Healpix_3.82/data/'
```

```python
kappa_LH = hp.read_map('./kappa_map_grid-768_nside-1024.fits')
kappa_OG = hp.read_map('../jax_implementation/output/kappa-map_websky1lpt_nside1024_768.fits')
```

```python
Cl_LH = hp.anafast(kappa_LH, lmax=1024, use_weights=True, datapath=hpx_datapath)
Cl_OG = hp.anafast(kappa_OG, lmax=1024, use_weights=True, datapath=hpx_datapath)
```

```python
Cl_OG[0], Cl_LH[0]
```

```python
ells = np.arange(1024+1)
plt.plot(ells[1:], 4e6*Cl_OG[1:], lw=1., label='without holes removed')
plt.plot(ells[1:], 4e6*Cl_LH[1:], lw=1., label='with holes removed')
# plt.ylim(-0.1, 1.4)
plt.xlabel(r'\ell')
plt.ylabel(r'$4\times 10^6$ $C^{\kappa \kappa}_\ell$')
plt.xscale('log')
plt.legend(frameon=False)
plt.yscale('log')
```
